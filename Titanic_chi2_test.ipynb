```python
import pandas as pd
df=pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.head()
```

```python

import numpy as np

def categorize_age(age):
    if pd.isnull(age):
        return np.nan
    elif age <= 1:
        return 'Младенцы'
    elif age <= 12:
        return 'Дети'
    elif age <= 17:
        return 'Подростки'
    elif age <= 29:
        return 'Молодежь'
    elif age <= 59:
        return 'Зрелый возраст'
    else:
        return 'Старики'

df['ВозрастКатегория'] = df['Age'].apply(categorize_age)
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['ВозрастКатегория'], df['Pclass'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Статистика χ²:", chi2)
print("Степени свободы:", dof)
print("p-значение:", p)
```

Следовательно, возрастная категория статистически значимо связана с классом пассажира на Титанике, так как p<0.05
